# **5.1 The Properties of Determinants**

- The determinant of a square matrix is a single scalar loaded with structural signal:
  - $\det A = 0$ ⇢ $A$ is non-invertible.
  - If $A$ is invertible, then $\det(A^{-1}) = \frac{1}{\det A}$.  
    Example: $\det A = 2 \Rightarrow \det(A^{-1}) = \tfrac12$.
  - Determinants give a closed-form for every entry of $A^{-1}$.

- Product of pivots gives the determinant:  
  Pivot product:  
  $a(d - \tfrac{c}{a} b) = ad - bc = \det A$

- Three computation pipelines for an $n \times n$ determinant:
  1. Multiply $n$ pivots (up to a $\pm$ sign) — **pivot formula**  
  2. Sum $n!$ signed terms — **big formula**  
  3. Combine $n$ smaller determinants — **cofactor formula**

- Sign flips matter:
  - Swapping two rows/columns flips the sign of the determinant.

- Baseline properties:
  1. Determinants give $A^{-1}$ and $A^{-1}b$ (Cramer's Rule).
  2. $|\det A|$ is the volume of the parallelepiped formed by rows/columns.
  3. For eigenvalues $\lambda$, $\det(A - \lambda I)=0$.


### **Properties of the Determinant**

1. **Identity matrix**  
   $$\det I_n = 1$$ 
   Example:  
   $$\begin{vmatrix}1 & 0 \\ 0 & 1\end{vmatrix} = 1$$

2. **Row swap = sign reversal**  
   $$\begin{vmatrix}c & d \\ a & b\end{vmatrix} = -\begin{vmatrix}a & b \\ c & d\end{vmatrix}$$

3. **Linearity in each row (others fixed)**  
   - Scaling row 1 by $t$:  
     $$\begin{vmatrix}ta & tb \\ c & d\end{vmatrix} = t \begin{vmatrix}a & b \\ c & d\end{vmatrix}$$
   - Adding rows:  
     $$\begin{vmatrix}a+a' & b+b' \\ c & d\end{vmatrix} = 
     \begin{vmatrix}a & b \\ c & d\end{vmatrix} +
     \begin{vmatrix}a' & b' \\ c & d\end{vmatrix}$$

4. **Equal rows ⇒ determinant zero**  
   $$\begin{vmatrix}a & b \\ a & b\end{vmatrix} = 0$$

5. **Row replacement doesn’t change $\det$**  
   (Elimination rule)  
   $$\begin{vmatrix}a & b \\ c - la & d - lb\end{vmatrix}
    = \begin{vmatrix}a & b \\ c & d\end{vmatrix}$$

6. **Zero row ⇒ determinant zero**  
   $$\begin{vmatrix}0 & 0 \\ c & d\end{vmatrix} = 0$$

7. **Triangular matrix ⇒ product of diagonal**  
   $$\det A = a_{11}a_{22}\cdots a_{nn}$$

8. **Singular ⇢ determinant 0**  
   - For $2 \times 2$:  
   $$\begin{vmatrix}a & b \\ c & d\end{vmatrix} = 0 \iff ad - bc = 0$$

9. **Product rule**  
   $$\det(AB) = (\det A)(\det B)$$  
   In particular:  
   $$A A^{-1} = I \Rightarrow (\det A)(\det A^{-1}) = 1$$

10. **Transpose invariance**  
    $$\det(A^T) = \det A$$



### **Column-wise implications**

(From $\det A = \det A^T$)
- Column swap flips sign.
- Zero or duplicate columns ⇒ determinant zero.
- Scaling a column scales the determinant.
- Determinant is linear in each column.


**Key Ideas**

1. Determinant is defined by:  
   - $\det I = 1$  
   - Row-swap sign flip  
   - Linearity in each row  
2. After elimination:  
   $\det A = \pm (\text{product of pivots})$
3. $\det A = 0$ ⇢ $A$ not invertible  
4. Two flagship identities:  
   - $\det(AB) = \det A \cdot \det B$  
   - $\det(A^T) = \det A$
