# **2.1 Vectors and Linear Equations**

- **Rows:** The row picture shows two lines meeting at a single point (the solution).  
- **Columns:** The column picture combines the column vectors on the left side to produce the vector $b$ on the right side.  
- **Typical Linear Algebra:** Look at a matrix by rows and columns. Its rows give the **row picture**, and its columns give the **column picture**. We combine those equations into a matrix problem $Ax = b$.


### **Four Steps to Understanding Elimination Using Matrices**

1. **Elimination:** Transform $A$ to a triangular matrix $U$ using a sequence of matrix steps $E_{ij}$.  
2. **Back Substitution:** Solve the triangular system from bottom to top.  
3. **LU Factorization:** In matrix language, $A$ is factored as $LU = (\text{lower triangular})(\text{upper triangular})$.  
4. **Invertibility:** Elimination succeeds if $A$ is invertible (may require row exchanges).


### **Three Equations in Three Unknowns**

Let there be three equations and three unknowns $x, y, z$:

$$
\begin{cases}
x + 2y + 3z = 6 \\
2x + 5y + 2z = 4 \\
6x - 3y + z = 2
\end{cases}
$$

- We look for numbers $x, y, z$ that solve all equations â€” a solution **might or might not exist**.  
- When the number of unknowns matches the number of equations, it usually has **one solution**.  

**Visualizations before solving:**

- **Row picture:** Three planes meeting at one point.  
- **Column picture:** Combines three columns to produce $b = (6, 4, 2)$.


### **The Matrix Form of the Equations**

- **Basic Question:** What does it mean to "multiply $A$ times $x$"? We can multiply by **rows** or by **columns**.

**Multiplication by rows:**

$$
Ax =
\begin{bmatrix}
\text{row}_1 \cdot x \\
\text{row}_2 \cdot x \\
\text{row}_3 \cdot x
\end{bmatrix}
$$

**Multiplication by columns:**

$$
Ax = x \cdot (\text{column}_1) + y \cdot (\text{column}_2) + z \cdot (\text{column}_3)
$$


**Key Ideas**

1. The basic operations on vectors are **scalar multiplication** $cv$ and **vector addition** $v + w$.  
2. Together, these operations give **linear combinations** $cv + dw$.  
3. **Matrix-vector multiplication $Ax$** can be computed by dot products (row by row), but must also be understood as a **combination of the columns of $A$**.  
4. **Column picture:** $Ax = b$ asks for a combination of columns to produce $b$.  
5. **Row picture:** Each equation in $Ax = b$ gives a line ($n = 2$), plane ($n = 3$), or "hyperplane" ($n > 3$). They intersect at the solution or solutions, if any.
