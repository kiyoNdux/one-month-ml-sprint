# **2.5 Inverse Matrices**

- **Definition:** The matrix $A$ is **invertible** if there exists a matrix $A^{-1}$ that "inverts" $A$:

$$
\text{Two-sided inverse: } A^{-1}A = I \quad \text{and} \quad AA^{-1} = I
$$

- Not all matrices have inverses. The first question for any square matrix is: *Is $A$ invertible?*

### **Six Notes about $A^{-1}$:**

**1.** The inverse exists **if and only if elimination produces $n$ pivots** (row exchanges are allowed). Elimination solves $Ax=b$ without explicitly using $A^{-1}$.  
**2.** The matrix $A$ cannot have **two different inverses**. Suppose $BA=I$ and $AC=I$, then $B=C$.  
**3.** If $A$ is invertible, the **only solution** to $Ax=b$ is

    $$
    x = A^{-1}b
    $$

- Multiply $Ax=b$ by $A^{-1}$: $x = A^{-1}Ax = A^{-1}b$. 

**4.** Suppose there exists a nonzero vector $x$ such that $Ax=0$. Then $A$ **cannot have an inverse**, because no matrix can bring $0$ back to $x$. If $A$ is invertible, then $Ax=0$ has only the zero solution: $x = A^{-1}0 = 0$.  
**5.** A $2 \times 2$ matrix is invertible if and only if $\det(A) = ad - bc \neq 0$.  
**6.** A diagonal matrix has an inverse **provided no diagonal entries are zero**.


### **The Inverse of a Product**

- If $A$ and $B$ are invertible, then so is $AB$:

$$
(AB)^{-1} = B^{-1} A^{-1}
$$

- Inverse order reverses for multiple matrices:

$$
(ABC)^{-1} = C^{-1} B^{-1} A^{-1}
$$


### **Calculating $A^{-1}$ by Gauss-Jordan Elimination**

- Gauss-Jordan: Multiply $[A \ I]$ by row operations to get $[I \ A^{-1}]$.


### **Singular versus Invertible**

- $A^{-1}$ exists **exactly** when $A$ has a full set of $n$ pivots.  
- Proof via elimination:

1. With $n$ pivots, elimination solves all equations $Ax_i = e_i$. The columns $x_i$ form $A^{-1}$. Then $AA^{-1}=I$.  
2. Elimination is a sequence of multiplications by $I$, $P$ (row exchanges), $D^{-1}$ (pivot division), and $E$ (elimination).  
3. If a right-inverse equals a left-inverse, $A$ is square and invertible.  

- **Why pivots are necessary (Reverse Reasoning):**

1. Without $n$ pivots, elimination produces a zero row.  
2. If $AC = I$ were possible, multiplying by elimination matrices $M$ would produce $MAC = M$, but the zero row prevents this.  
3. Therefore, $A$ must have $n$ pivots for invertibility.  


### **Recognizing an Invertible Matrix**

- **Diagonally dominant matrices** are invertible.  
- Each diagonal entry dominates its row:

$$
|a_{ii}| > \sum_{j \neq i} |a_{ij}|
$$


**Key Ideas**

1. The inverse matrix satisfies $AA^{-1}=I$ and $A^{-1}A=I$.  
2. $A$ is invertible **iff** it has $n$ pivots (row exchanges allowed).  
3. If $Ax=0$ has a nonzero solution, $A$ is **not invertible**.  
4. The inverse of a product reverses order: $(AB)^{-1} = B^{-1}A^{-1}, (ABC)^{-1} = C^{-1}B^{-1}A^{-1}$.  
5. Gauss-Jordan finds $A^{-1}$ by row-reducing $[A \ I] \to [I \ A^{-1}]$.  
6. Diagonally dominant matrices are invertible.
