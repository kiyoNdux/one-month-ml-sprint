# **4.2 Projections**

- When $b$ is projected onto a line, its projection $p$ is the part of $b$ along that line.  
  If $b$ is projected onto a plane, $p$ is the part in that plane. The projection is $p = Pb$.

- The projection matrix $P$ multiplies $b$ to give $p$.

- Our problem: project any $b$ onto the column space of any $m \times n$ matrix.


### **Projection Onto a Line**

- A line goes through the origin in the direction of $a = (a_1, \dots, a_m)$.  
  We want a point $p$ closest to $b = (b_1, \dots, b_m)$ along that line.

- **Key idea:** Orthogonality — the line from $b$ to $p$ is perpendicular to $a$.

- **Computing $p$ algebraically:**
  **1.** The projection $p$ is some multiple of $a$:  
     $p = \hat{x} a$  
  **2.** Compute $\hat{x}$ using orthogonality.  
     Error $e = b - \hat{x} a$ is perpendicular to $a$:  
     $$
     a^T (b - \hat{x} a) = 0 \quad \Rightarrow \quad \hat{x} = \frac{a^T b}{a^T a}
     $$
  **3.** Projection vector: $p = \hat{x} a = \frac{a^T b}{a^T a} a$  
     Projection matrix: $P = \frac{aa^T}{a^T a}$

- **Special cases:**
  - $b = a \Rightarrow \hat{x} = 1, \; p = a$
  - $b \perp a \Rightarrow a^T b = 0, \; p = 0$

- **Properties of $P$:**
  - $P$ is rank one: column $a$ times row $a^T$
  - Doubling $a$ does not change $P$
  - $P^2 = P$
  - $I - P$ projects onto the perpendicular subspace (the error $e$)


### **Projection Onto a Subspace**

- Let $a_1, \dots, a_n \in \mathbb{R}^m$ be linearly independent.  
  Find $p = \hat{x}_1 a_1 + \dots + \hat{x}_n a_n = A \hat{x}$ closest to $b$.

- **Computing $p$:**
  1. Error vector: $e = b - A \hat{x}$  
     Orthogonal to all $a_i$: $(a_i)^T e = 0$ for $i = 1, \dots, n$
  2. Stack into matrix form: $A^T (b - A \hat{x}) = 0$
  3. Solve the normal equation:  
     $$
     A^T A \hat{x} = A^T b
     $$
  4. Solution:  
     $$
     \hat{x} = (A^T A)^{-1} A^T b, \quad p = A \hat{x} = A (A^T A)^{-1} A^T b
     $$
  5. Projection matrix:  
     $$
     P = A (A^T A)^{-1} A^T, \quad p = Pb
     $$

- **Comparison with line projection:**  
  - Single column $A = a \Rightarrow A^T A = a^T a$  
  - $\hat{x} = a^T b / a^T a$, $p = a (a^T b / a^T a)$, $P = aa^T / a^T a$

- **Key points:**
  1. The subspace is the column space of $A$.
  2. Error $e = b - A \hat{x}$ is perpendicular to the column space → $e \in N(A^T)$.
  3. Projection forms a right triangle: sides $p, e, b$.
  4. $A^T A$ is invertible iff $A$ has linearly independent columns.
  5. Projection matrix $P$ is symmetric ($P^T = P$) and idempotent ($P^2 = P$).
  6. Distance from $b$ to subspace $C(A)$: $\|e\|$.


**Key Ideas**

1. Projection of $b$ onto line through $a$:  
   $$
   p = a \hat{x} = a \frac{a^T b}{a^T a}
   $$

2. Rank-one projection matrix: $P = \frac{aa^T}{a^T a}$
3. Projecting $b$ onto a subspace leaves $e = b - p$ perpendicular to the subspace.
4. Full-rank $A$: solve $A^T A \hat{x} = A^T b$ to get $\hat{x}$, then $p = A \hat{x}$
5. Projection matrix for subspace: $P = A (A^T A)^{-1} A^T$, satisfies $P^T = P$ and $P^2 = P$, and $Pb = p$.
