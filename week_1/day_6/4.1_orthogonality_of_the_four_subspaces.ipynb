# **4.1 Orthogonality of the Four Subspaces**

- **Orthogonal vectors:** 
  - $v^T w = 0$  
  - $\|v\|^2 + \|w\|^2 = \|v + w\|^2$

- The row space is perpendicular to the nullspace.

- The column space is perpendicular to the nullspace of $A^T$.

**Definition:**  
Two subspaces $V$ and $W$ are orthogonal if every vector $v \in V$ is perpendicular to every vector $w \in W$.

- Orthogonal subspaces: $v^T w = 0$ for all $v \in V$ and $w \in W$.
- Every vector $x$ in the nullspace is perpendicular to every row of $A$, because $Ax = 0$.
- The nullspace $N(A)$ and the row space $C(A^T)$ are orthogonal subspaces of $\mathbb{R}^n$.

Matrix view:

$$
Ax =
\begin{bmatrix}
\text{row}_1 \\
\vdots \\
\text{row}_m
\end{bmatrix}
x =
\begin{bmatrix}
0 \\
\vdots \\
0
\end{bmatrix}
$$

- Every vector $y$ in the nullspace of $A^T$ is perpendicular to every column of $A$.
- The left nullspace $N(A^T)$ and the column space $C(A)$ are orthogonal in $\mathbb{R}^m$.

$$
A^T y =
\begin{bmatrix}
(\text{column}_1)^T \\
\vdots \\
(\text{column}_n)^T
\end{bmatrix}
y =
\begin{bmatrix}
0 \\
\vdots \\
0
\end{bmatrix}
$$

- The dot product of $y$ with every column of $A$ is zero, so $y$ is perpendicular to the entire column space.

    
### **Orthogonal Complements**

- The fundamental subspaces are not only orthogonal pairwise — their dimensions also fit perfectly.

**Definition:**  
The orthogonal complement of a subspace $V$ is the set of all vectors perpendicular to $V$.  
It is denoted by $V^\perp$.

- By definition, the nullspace $N(A)$ is the orthogonal complement of the row space $C(A^T)$:  
  - Every $x$ perpendicular to the rows satisfies $Ax = 0$.
- The reverse is also true: if $x$ is orthogonal to the nullspace, it lies in the row space.

**Fundamental Theorem of Linear Algebra — Part 2**

- $N(A)$ is the orthogonal complement of $C(A^T)$ in $\mathbb{R}^n$  
- $N(A^T)$ is the orthogonal complement of $C(A)$ in $\mathbb{R}^m$

**Decomposition:**  
Every vector $x$ in $\mathbb{R}^n$ can be written as

$$
x = x_r + x_n,
$$

where:

- $x_r$ is the row space component  
- $x_n$ is the nullspace component  

Under multiplication:

- $A x_n = 0$  
- $A x_r = Ax$ (lands in the column space)


### **Combining Bases from Subspaces**

**Bases of $\mathbb{R}^n$ facts:**

- Any $n$ independent vectors in $\mathbb{R}^n$ form a basis.
- Any $n$ spanning vectors in $\mathbb{R}^n$ are independent.

**Two facts about square invertible matrices:**

- If the $n$ columns of $A$ are independent → they span $\mathbb{R}^n$ → $Ax = b$ is solvable.
- If the $n$ columns span $\mathbb{R}^n$ → they are independent → $Ax = b$ has a unique solution.

**Longer facts:**

- Uniqueness $\Rightarrow$ existence, and existence $\Rightarrow$ uniqueness.  
  - If there are no free variables, the solution is unique → $A$ is invertible.
- If $Ax = b$ is solvable for every $b$:  
  - Elimination produced no zero rows  
  - There are $n$ pivots  
  - The nullspace contains only $x = 0$ (unique solution)
- With bases for the row space and nullspace, we get $r + (n - r) = n$ vectors.  
  These vectors are independent → they span $\mathbb{R}^n$.
- Each $x$ splits as $x = x_r + x_n$.


**Key Ideas**

1. Subspaces $V$ and $W$ are orthogonal if every $v \in V$ is orthogonal to every $w \in W$.

2. $V$ and $W$ are orthogonal complements if each contains all vectors perpendicular to the other.  
   In $\mathbb{R}^n$,  
   $$
   \dim(V) + \dim(W) = n.
   $$

3.  
   - $N(A)$ and $C(A^T)$ are orthogonal complements with dimensions $(n-r) + r = n$.  
   - $N(A^T)$ and $C(A)$ are orthogonal complements in $\mathbb{R}^m$ with $(m-r) + r = m$.

4. Any $n$ independent vectors in $\mathbb{R}^n$ span it, and any $n$ spanning vectors are independent.
