# **1.1 Vectors and Linear Combinations**

**Analogy:** “You can't add apples and oranges,” that's why  
  we make two separate pairs of numbers `v_1` and `v_2`.  
  The pair produces a 2D vector `v`.

- `v_1` = the first component of `v`
- `v_2` = the second component of `v`
- We write `v` as a **column**, not a row


### **Vector Addition**

- Let **v** and **w** be 2D vectors:
- **v** = (v₁, v₂)  
- **w** = (w₁, w₂)
- **v + w** = (v₁ + w₁, v₂ + w₂)


```python
def add_vectors(u, v):
    res = []
    for index_u, value_u in enumerate(u):
        for index_v, value_v in enumerate(v):
            if index_u == index_v:
                res.append(value_u + value_v)
    return res
        

u = [1, 2]
v = [-1, 2]
add_vectors(u ,v)
```

### **Scalar Multiplication**

- Let **c = 2** and **v = (v₁, v₂)**:  
  **2v = (2v₁, 2v₂)**

- Let **c = -1**:  
  **−v = (−v₁, −v₂)**


```python
def scale_vectors(c, v):
    res = []
    for i in v:
        res.append(c*i)
    return res

c = 2
v = [1, 2]
scale_vectors(c, v)
```

### **Linear Combinations**

- Combining addition with scalar multiplication produces a **linear combination**.
- Example: the sum of **cv** and **dw** is a linear combination **cv + dw**.

**Four special linear combinations:**

- **1v + 1w** = sum of vectors  
- **1v − 1w** = difference of vectors  
- **0v + 0w** = zero vector  
- **cv + 0w** = vector **cv** in the direction of **v**


```python
def linear_comb(c, v, d, w):
    scaled_v =[]
    scaled_w = []
    res = []

    for i in v:
        scaled_v.append(c*i)

    for j in w:
        scaled_w.append(d*j)

    for index_v, value_v in enumerate(scaled_v):
        for index_w, value_w in enumerate(scaled_w):
            if index_v == index_w:
                res.append(value_v + value_w)
    return res

c = 1
v = [1, 2, 3]
d = 2
w = [4, 5, 6]
linear_comb(c, v, d, w)
```

### **Vector Representations**

In algebra, we can represent a 2D vector **v** as:

- two numbers  
- an arrow from **(0, 0)**  
- a point in the plane  

We add using the numbers, and we visualize **v + w** using arrows:

- at the end of **v**, place the start of **w**


```python
import matplotlib.pyplot as plt
import numpy as np

# Define vectors
v = np.array([2, 3])
w = np.array([1, 2])

# Origin
origin = np.array([0, 0])

# Plot vectors as arrows
plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1, color='r', label='v')
plt.quiver(*origin, *w, angles='xy', scale_units='xy', scale=1, color='b', label='w')

# Vector addition
v_plus_w = v + w
plt.quiver(*origin, *v_plus_w, angles='xy', scale_units='xy', scale=1, color='g', label='v + w')

# Set plot limits
plt.xlim(0, 5)
plt.ylim(0, 6)
plt.grid()
plt.legend()
plt.show()
```

### **Vectors in Three Dimensions**

- A vector with 2 components corresponds to a point in the **xy-plane**,  
  a vector with 3 components corresponds to a point in **xyz space**.

- The arrow always starts at the **origin**, where all the components  
  of the vector are zeros — this is where the axes meet.


```python
from mpl_toolkits.mplot3d import Axes3D

# Define vectors
u = np.array([1, 2, 1])
v = np.array([2, 1, 0])
w = np.array([0, 1, 2])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot vectors from origin
ax.quiver(0, 0, 0, *u, color='r', label='u')
ax.quiver(0, 0, 0, *v, color='b', label='v')
ax.quiver(0, 0, 0, *w, color='g', label='w')

# Axes limits
ax.set_xlim([0,3])
ax.set_ylim([0,3])
ax.set_zlim([0,3])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.show()
```

### **The Important Questions**

- For one vector **u**, the only linear combinations are the multiples **cu**.  
- For two vectors, the combinations are **cu + dv**.  
- For three vectors, the combinations are **cu + dv + ew**.

Suppose the vectors **u**, **v**, and **w** are in 3D space:

1. *“What is the picture of all combinations **cu**?”*  
2. *“What is the picture of all combinations **cu + dv**?”*  
3. *“What is the picture of all combinations **cu + dv + ew**?”*

- The answer depends on the particular vectors **u**, **v**, and **w**.  
  If they were zero, then every combination would be zero.

- If they are typical nonzero vectors, then:  
  1. The combinations fill a **line** through (0,0,0).  
  2. The combinations fill a **plane** through (0,0,0).  
  3. The combinations fill **all 3D space**.

- There are also cases where, even though we have 3 components and 3 vectors,  
  if one is **dependent**, then it produces only a **2D plane**.


**Note**

- A **zero vector** is not the same as the number **zero**.  
- Every time we see a space of vectors, the **zero vector is always included**.  
- **w + v** gives the same answer as **v + w**.  
- The vector **(x, y)** ≠ **(x, y, 0)** — the first is in 2D, the second is in 3D.


**Key Ideas**

1. A vector in 2D space has two components **v₁** and **v₂**.  
2. **v + w = (v₁ + w₁, v₂ + w₂)** and **cv = (cv₁, cv₂)** are computed **component-wise**.  
3. A linear combination of three vectors **u**, **v**, and **w** is **cu + dv + ew**.  
4. Take all linear combinations of **u**, or **u and v**, or **u, v, w**.  
   In three dimensions, these combinations typically fill a **line**, **plane**,  
   then the **whole space** **ℝ³**.
