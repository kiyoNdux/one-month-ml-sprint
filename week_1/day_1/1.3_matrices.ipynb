# **1.3 Matrices**

### **Linear Combinations and Matrix Representation**

- Let the three vectors be:

$$
u = \begin{bmatrix} 1 \\ -1 \\ 0 \end{bmatrix}, \quad
v = \begin{bmatrix} 0 \\ 1 \\ -1 \end{bmatrix}, \quad
w = \begin{bmatrix} 0 \\ 0 \\ 1 \end{bmatrix}
$$

- Their **linear combinations** in 3D space are:

$$
x_1 u + x_2 v + x_3 w =
x_1 \begin{bmatrix} 1 \\ -1 \\ 0 \end{bmatrix} +
x_2 \begin{bmatrix} 0 \\ 1 \\ -1 \end{bmatrix} +
x_3 \begin{bmatrix} 0 \\ 0 \\ 1 \end{bmatrix}
$$

- Rewriting as a **matrix equation**:

$$
A x = 
\begin{bmatrix}
1 & 0 & 0 \\
-1 & 1 & 0 \\
0 & -1 & 1
\end{bmatrix}
\begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix}
=
\begin{bmatrix}
x_1 \\
x_2 - x_1 \\
x_3 - x_2
\end{bmatrix}
= b
$$

- Here:
  - **Matrix A** acts on the vector **x**  
  - The output **Ax** is a combination **b** of the columns of **A**:

$$
b = 
\begin{bmatrix} b_1 \\ b_2 \\ b_3 \end{bmatrix} =
\begin{bmatrix} x_1 \\ x_2 - x_1 \\ x_3 - x_2 \end{bmatrix}
$$

- **Input:** $x = [x_1, x_2, x_3]^T$
- **Output:** $b = Ax$

### **Matrix Multiplication: Row by Row**

- Consider the matrix-vector multiplication:

$$
A x =
\begin{bmatrix}
1 & 0 & 0 \\
-1 & 1 & 0 \\
0 & -1 & 1
\end{bmatrix}
\begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix}
$$

- This can be computed **row by row**:

$$
\begin{bmatrix}
(1, 0, 0) \cdot (x_1, x_2, x_3) \\
(-1, 1, 0) \cdot (x_1, x_2, x_3) \\
(0, -1, 1) \cdot (x_1, x_2, x_3)
\end{bmatrix}
=
\begin{bmatrix}
x_1 \\
x_2 - x_1 \\
x_3 - x_2
\end{bmatrix}
$$

- Each row of **A** is **dotted** with the vector **x** to produce the corresponding entry in **Ax**.


### **Linear Equations**

- **Old question:** "Compute the linear combination $x_1 u + x_2 v + x_3 w$ to find $b$"  
- **New question:** "Which combination of $u, v, w$ produces a particular $b$?"

Equations $(Ax = b)$:

$$
\begin{aligned}
x_1 &= b_1 \\
-x_1 + x_2 &= b_2 \\
-x_2 + x_3 &= b_3
\end{aligned}
$$

Solution $(x = A^{-1}b)$:

$$
x_1 = b_1, \quad
x_2 = b_1 + b_2, \quad
x_3 = b_1 + b_2 + b_3
$$

- If:

$$
b = \begin{bmatrix} 0 \\ 0 \\ 0 \end{bmatrix} \quad \text{then} \quad 
x = \begin{bmatrix} 0 \\ 0 \\ 0 \end{bmatrix}
$$

- This shows the idea that if the output $b = 0$, then the input must be $x = 0$.  
- This isn’t always the case for all matrices.  
- This matrix is also **invertible**. From $b$ we can recover $x$. We write $x$ as $A^{-1}b$ (both ideas are shown here just for exposure).


### **The Inverse Matrix**

- The equation we used above tells us **two important facts**:  
1. For every $b$, there is **one solution** to $Ax = b$
2. The matrix $A^{-1}$ produces $x = A^{-1}b$

- Note that this is only true for **invertible matrices**. Now we ask:  
1. Is there a solution?  
2. If yes, how do we find it?


### **Cyclic Differences**

- Let the three vectors be:

$$
u = \begin{bmatrix} 1 \\ -1 \\ 0 \end{bmatrix}, \quad
v = \begin{bmatrix} 0 \\ 1 \\ -1 \end{bmatrix}, \quad
w = \begin{bmatrix} -1 \\ 0 \\ 1 \end{bmatrix}
$$

- Rewriting as a matrix:

$$
C x =
\begin{bmatrix}
1 & 0 & -1 \\
-1 & 1 & 0 \\
0 & -1 & 1
\end{bmatrix}
\begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix}
=
\begin{bmatrix}
x_1 - x_3 \\
x_2 - x_1 \\
x_3 - x_2
\end{bmatrix} = b
$$

**Case 1:**  

$$
b = \begin{bmatrix} 0 \\ 0 \\ 0 \end{bmatrix}
$$

- In this case, there are **infinitely many solutions**.

**Case 2:**  

$$
b = \begin{bmatrix} 1 \\ 3 \\ 5 \end{bmatrix} \quad \text{or} \quad b = \begin{bmatrix} c \\ c \\ c \end{bmatrix}
$$

- In this case, there is **no solution**.

- **Main points to introduce these ideas:**
1. The equation has **1 solution** → matrix is square and invertible.  
2. The equation has **infinitely many solutions** → there exists a nonzero $x$ that produces 0, meaning one vector is **dependent**.  
3. The equation has **no solution** → $b$ is not in the **column space**.


### **Independence and Dependence**

- **Just a look at the idea:**
  1. **Independence:** $u, v, w$ are independent if **no combination except** $0u + 0v + 0w = 0$ gives $b = 0$  
  2. **Dependence:** $u, v, w$ are dependent if **other combinations** like $u + v + w$ give $b = 0$

- **Independent columns:** $Ax = 0$ has **one solution** → $A$ is an **invertible matrix**  
- **Dependent columns:** $Cx = 0$ has **many solutions** → $C$ is a **singular matrix**


**Key Ideas**

1. **Matrix times vector:** $Ax$ = combination of the columns of $A$
2. **Solution to Ax = b:** $x = A^{-1}b$, when $A$ is invertible  
3. **Cyclic matrix:** no inverse; its three columns lie in the **same plane**. Dependent columns sum to the zero vector → $Cx = 0$ has many solutions
